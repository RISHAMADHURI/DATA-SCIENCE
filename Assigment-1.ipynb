```python
import pandas as pd
```

```python
students = pd.read_csv(r"/content/students.csv")
students
```

```python
details = pd.read_csv(r"/content/details.csv")
details
```

```python
#Inner Join
merged_df=pd.merge(students,details,on="StudentID",how="inner")
merged_df
```

```python
#Left Join
merged_df1=pd.merge(students,details,on="StudentID",how="left")
merged_df1
```

```python
#Right Join
merged_df2=pd.merge(students,details,on="StudentID",how="right")
merged_df2
```

```python
#Outer Join
merged_df3=pd.merge(students,details,on="StudentID",how="outer")
merged_df3
```

```python
merged_df3.set_index("StudentID", inplace=True)
merged_df3
```

```python
merged_df3.to_csv("merged_students_details.csv" , index=False)
```

```python
merged_students_details=pd.DataFrame(merged_df3)
```
